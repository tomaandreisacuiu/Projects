The COVID pandemic has changed our lives. While we may feel that the changes are often for the worse, there are also indications that some aspects of life may improve. <div class="alert alert-block alert-danger">
    <b>Note on proxies:</b>    
    There is an underlying subcurrent beneath this assignment that runs very deep, and its significance extends far beyond this assignment. The <em>lockdown stringency index</em> that is computed is merely a <b>proxy</b> (or stand-in, replacement) for the stringency of the lockdown. As every proxy, it merely approximates <b>some</b> aspects of the concept it tries to capture (in this case, how severe a lockdown is), but it certainly doesn't capture them all and it undoubtedly has its inaccuracies. Put bluntly, <b>a proxy is always wrong</b>. Even the numbers for traffic intensity that we use in this assignment are merely a proxy, and though much closer to the actual thing, they don't fully capture the true traffic intensity in the Netherlands. Therefore, proxies always require a balanced view: they may be useful for some things, sometimes they can help you predict some variables (and at other times that will go wrong), but wrong use of proxies can lead to disasters. For further reading on this topic, we recommend Cathy O'Neil's book <em>Weapons of Math Destruction</em>.
</div>

### Read in the data

We first read in the data, parsing the relevant columns as dates and converting the times from UTC to Amsterdam time. Note that you do not need to understand the code below, but we describe it here in case you are interested.

````python
df_oxford = pd.read_csv('datasets/OxCGRT_NL.csv', parse_dates=['Date'])
df_oxford['Date'] = df_oxford['Date'].dt.tz_localize('Europe/Amsterdam')
df_oxford = df_oxford.set_index('Date')

df_data = pd.read_csv('datasets/traffic.csv', parse_dates=[0], index_col=[0])
df_data['average_traffic'] = df_data.mean(axis=1)
df_data.index = df_data.index.tz_convert('Europe/Amsterdam')
df_data.index.name = 'local_time'
df_data=df_data.iloc[0:-1]
df_data['stringency_index'] = df_oxford['StringencyIndex']
df_data['stringency_index'] = df_data['stringency_index'].fillna(method='ffill')
df_data['stringency_index'] = df_data['stringency_index'].fillna(0)
df_data.head()
````

Now we add some time features that will be useful in the exploration later.

````python
df_data['month'] = df_data.index.month
df_data['weekday'] = df_data.index.weekday
df_data['hour'] = df_data.index.hour
df_data['day_of_year'] = df_data.index.dayofyear
df_data['week'] = df_data.index.isocalendar().week.astype(int)
df_data['ISO week-counting year'] = df_data.index.isocalendar().year.astype(int)
df_data['helper'] = np.abs(df_data['week'].diff().fillna(0).apply(lambda _ : np.max([_, -1]))).astype(int)
df_data['week_since_2015'] = 1 + df_data['helper'].cumsum()
df_data['week_since_2015'] = df_data['week_since_2015'].fillna(1).astype(int)
df_data.drop('helper', axis=1, inplace=True)
for col in ['week_since_2015', 'ISO week-counting year']:
    df_oxford[col] = df_data[col]
df_data.head()
````
